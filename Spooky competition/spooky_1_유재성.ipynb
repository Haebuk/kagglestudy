# Natural Language Processing
거의 모든 자연어 처리 작업에서, 일반적으로 원시 텍스트를 형식으로 변환하기 위해 몇가지 사전 처리 단계를 거쳐야 합니다. 랜덤포레스트 모델에 단어 한 단락을 입력시키고 어느 작가로부터 왔는지 즉시 예측할 수 있으리라 기대할 수 없습니다. 텍스트 사전 처리는 다음과 같은 몇 가지 간단한 단계로 요약할 수 있습니다.

1. __Tokenization__: 텍스트를 개별 구성 단어로 구분합니다.
2. __Stopwords__: 발생 빈도가 너무 높으면 관련 텍스트를 탐지하는 데 도움이 되지 않으므로 너무 자주 발생하는 단어를 모두 버립니다. (아주 드물게 발생하는 단어를 버리는 것도 고려하십시오.)
3. __Stemming__: 동일한 의미를 전달하는 다양한 단어들을 결합합니다.
4. __Vectorization__: 텍스트를 벡터 형식으로 변환합니다. 가장 간단한 방법 중 하나는 bag-of-words 접근입니다. 여기서 행렬(말뭉치의 각 문서 또는 텍스트)을 만듭니다. 가장 간단한 형태로, 이 행렬은 단어 빈도(단어 수)를 저장하며 종종 원시 텍스트의 벡터화라고 합니다.

__Natural Language Toolkit(NLTK)__: 자연어 처리를 보다 편리하기 위해 NLP에서 가장 유용한 툴킷 중 하나인 NLTK 모듈을 소개하겠습니다. 툴킷을 가져오는 방법은 다음과 같습니다.

```python
import nltk
```

## Tokenization

```python
# 첫 번째 텍스트 요소를 문자열로 저장
first_text = train.text.values[0]
print(first_text)
print('='*90)
print(first_text.split(' '))
```

토큰화를 처음 시도했을 때 알 수 있듯이, 문장을 개별 요소로 구분하는 것이 완전히 정확한 것은 아닙니다. 예를 들어, 'process,'라는 용어가 포함된 두 번째 요소를 살펴보겠습니다. 쉼표도 포함되었으며 'process'라는 단어와 함께 용어 자체로 처리되고 있습니다. 이상적으로는 쉼표와 단어가 목록의 서로 다른 두 개의 요소에 포함되기를 원합니다. 

순수한 파이썬 리스트 작업으로 수행하는 것은 복잡하기 때문에 NLTK 라이브러리를 사용합니다. 'word_tokenize()'메서드를 사용하여 단일 단어와 구두점을 자동으로 개별 요소로 제거합니다.

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
first_text_list = nltk.word_tokenize(first_text)
print(first_text_list)
```

## Stopword Removal
일반적으로 너무 흔하고 높은 빈도로 나타나는 단어들은 학습 프로세스에 크게 기여하지 않습니다. 중지 단어에는 'to' 또는 'the' 같은 용어가 포함되어 있으므로 사전 처리 단계에서 제거하는 것이 유리합니다. NLTK는 사전 정의된 153개의 영어 중지 단어 목록이 제공됩니다.

```python
stopwords = nltk.corpus.stopwords.words('english')
len(stopwords)
```

토큰화된 단어 목록에서 중지 단어를 필터링하려면 다음과 같은 리스트 컴프리헨션을 사용하면 됩니다.

```python
first_text_list_cleaned = [word for word in first_text_list if word.lower() not in stopwords]
print(first_text_list_cleaned)
print('='* 90)
print('Length of original list: {0} words\n'
      'Length of list after stopwords removal: {1} words'
      .format(len(first_text_list), len(first_text_list_cleaned)))
```

보시다시피, 중지 단어가 제거된 목록은 'I', 'me', 'to', 'the'와 같이 일반적으로 사용되는 단어가 제거되어 전보다 상당히 짧아졌습니다.

## Stemming and Lemmatization
중지 단어를 제거한 후, NLP의 다음 단계는 Stemming의 과정입니다. 이 단계에서는 유사한 단어들을 하나로 줄이려고 합니다. 만약 우리가 'run', 'runs', 'running'을 가지고 있다면, 세 개로 구별되는 단어들이 'run'으로 묶이는 것을 원합니다. (그러나 시제에 대한 정보를 잃게 됩니다.)

NLTK는 lancaster stemmer, stemming 알고리즘, Snowball stemming과 같은 다양한 스테머를 제공합니다. 다음 예는 porter stemmer 인스턴스를 만듭니다.

```python
stemmer = nltk.stem.PorterStemmer()
```

이제 스테머를 사용하여 'run'으로 줄일 수 있는지 확인할 수 있습니다.

```python
print('The stemmed form of running is: {}'.format(stemmer.stem('running')))
print('The stemmed form of runs is: {}'.format(stemmer.stem('runs')))
print('The stemmed form of run is: {}'.format(stemmer.stem('run')))
```

위와 같이 주어진 단어를 성공적으로 기본 형태로 줄였고 이것은 학습 및 분류 작업에서 큰 도움이 될 것입니다.

그러나 결과에 한 가지 결함이 있습니다. 그것은 특정 단어를 사람이 인식할 수 있는 기본 형태로 줄이기 위해 단어를 자르면서 꽤 조잡한 경험적 발견을 수반한다는 사실입니다. 따라서 이 프로세스는 단어를 자를 때 어휘 또는 단어 형식을 고려하지 않습니다. 다음 예를 보시죠.

```python
print('The stemmed form of leaves is: {}'.format(stemmer.stem('leaves')))
```

### Lemmatization to the rescue
그러므로 우리는 stemming의 다른 방법을 살펴보는데 이를 lemmatization이라고 합니다. 스테머와 달리, 데이터 세트를 레미타이징하는 것은 실제 사전이나 어휘를 기반으로 단어를 줄이는 것을 목표로 하므로 어휘적 의미를 갖지 않는 단어들을 줄이지 않습니다.

여기서는 NLTK를 활용하여 레미타이저를 초기화하고 다음과 같이 단어를 축소하는 방법을 살펴볼 수 있습니다.

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
print('The lemmatized form of leaves is: {}'.format(lemm.lemmatize('leaves')))
```

레미타이저가 작동하면서 단어를 어휘적으로 더 적합하게 변형한다는 것을 알게 되었습니다.

## Vectorizing Raw Text
방대한 NLP 문헌에는 원시 텍스트를 분석하기 위한 다양한 목적이 쓰여 있습니다. 어떤 경우에는 한 텍스트 본문의 유사성을 다른 텍스트 본문과 비교(클러스터링 기법/거리 측정), 텍스트 분류 및 본문을 구성하는 주제를 파악하고자 합니다. 

우리는 이제 기계 학습 모델에 원시 텍스트를 주입해야합니다. 토큰화, 중지 단어 제거 및 스테밍(또는 레미타이제이션)에 논의했으므로 더 깨끗한 텍스트 데이터 세트에 도달했습니다. 그러나 아직 기계는 읽지 못합니다. 기계는 비트와 숫자를 읽을 수 있으므로 먼저 텍스트를 숫자로 변환하여 bag-of-words로 알려진 매우 일반적인 접근 방식을 사용해야 합니다.

### The bag of words approach
이 접근 방식은 단어 수를 시작 블록으로 사용하고, 해당 특정 단어의 특정 벡터에 각 단어의 발생을 기록합니다. 'I love to eat Burgers', 'I love to eat Fries' 두 문장을 예로 들면, 먼저 [I, love, to, eat, Burgers, Fries]라는 단어의 카운트를 얻기 위해 토큰화합니다.

bag-of-words 접근 방식을 통해 텍스트를 벡터화하면 각 단어에 대해 6개의 개별 벡터를 얻게 됩니다. 이제 텍스트 대신 숫자로 구성된 행이 있기 때문에 어떤 열(피쳐)이 형성되는지 살펴봅니다. 이제 각 단어는 새롭게 변환덴 데이터 세트의 개별 피쳐가 됩니다. 이 점을 설명하기 위해 사이킷런 라이브러리를 활용하여 CountVectorizer 메서드를 통해 다음과 같이 단어 수의 벡터를 생성합니다.

```python
# 문장 정의
sentence = ['I love to eat Burgers',
            'I love to eat Fries']
vectorizer = CountVectorizer(min_df=0)
sentence_transform = vectorizer.fit_transform(sentence)
```

### Fitting the vectorizer to the dataset
여기서는 CounterVectorizer 함수를 통해 간단히 용어 빈도수의 개체를 초기화하고 생성합니다. 명시적으로 제공한 매개 변수(나머지는 기본값)는 최소값입니다. 여기서 매개변수 'min_df'는 최소 문서 빈도를 나타내며 해당 값보다 적게 발생하는 모든 단어를 삭제합니다. 

마지막으로 fit_transform 방법은 실제로 두 단계로 구성됩니다. 첫 번째 단계는 벡터라이저가 제공된 데이터 세트에 매핑되는 적합 방법입니다. 이 작업이 완료되면 아래와 같이 원시 텍스트가 벡터 형태로 변환되는 변환 방법을 통해 실제 벡터화 작업이 수행됩니다.

# Topic modeling
최종 목적지에 도착하여 다음과 같은 두 가지 주제 모델링 기법을 구현합니다.
1. Latent Dirichlet Allocation: 각 주제가 각 단어에 서로 다른 확률 가중치를 할당하는 말뭉치의 단어에 가중치를 할당하여 데이터 세트에 잠재된 주제를 발견하는 확률적 생성 모델입니다.
2. Non-negative Matrix Factorization: 입력 행렬을 가져와 행렬의 값이 음수가 아닌 것을 주의하여 이 행렬의 인자화를 다른 두 행렬로 근사화 하는 근사 기법입니다.

```python
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = '\nTopic #{}:'.format(index)
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words -1 :-1]])
        print(message)
        print('='*70)
```

## Putting all the preprocessing steps together
이제 이전 섹션에서 언급한 모든 사전 처리 단계를 수행할 수 있습니다. 

CounterVectorizer를 사용하여 원시 텍스트를 벡터화시키면 토큰화 및 중지 단어 필터링의 이중 단계가 자동으로 포함됩니다. 

벡터라이저에는 내장형 레미타이저가 없어 이를 구현해야하나, 다행히 CounterVectorizer 클래스를 확장하여 'build_analyzer'을 추가하면 됩니다.

### Extending the CountVectorizer class with a lemmatizer

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
```

여기서 객체 지향 프로그래밍의 몇가지 개념을 사용했습니다. 기본적으로 원본 사이킷런 CountVectorizer 클래스를 상속 및 하위 분류를 했으며 원시 텍스트 매트릭스의 각 목록에 대한 레미타이저를 구현하여 build_analyzer 메서드를 덮어썼습니다.

```python
# 전체 훈련 텍스트를 리스트에 저장
text = list(train.text.values)
# 카운트 벡터라이저 호출
tf_vectorizer = LemmaCountVectorizer(max_df=0.95,
                                     min_df=2,
                                     stop_words='english', 
                                     decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
```

## Latent Dirichlet Allocation
마지막으로 주제 모델링과 두 가지 비지도 학습 알고리즘의 구현에 대한 주제입니다.
사이킷런의 구현을 통해 LDA를 사용해보겠습니다.

### 말뭉치 - 문서 - 단어: 주제 생성
LDA에서 모델링 프로세스는 텍스트 말뭉치, 문서의 모음 D, 문서의 단어 W라는 세 가지를 중심으로 이루어집니다. 따라서 알고리즘은 이 말뭉치에서 주제 K를 다음과 같이 밝혀냅니다.
![title](./lda_model_graph.png)

βk 가 제공하는 Dirichlet 사전 분포를 통한 모델의 각 토픽 κ:

α로 매개 변수화된 또 다른 Dirichlet 배포판으로 각 문서를 모델링:

이어서 문서화된 경우, 다항 분포를 통해 주제를 생성하고 이를 역추적하여 다른 다항 분포를 통해 해당 주제와 관련된 단어를 생성하는데 사용합니다.

LDA 알고리즘은 먼저 주제의 혼합 모델을통해 문서를 모델링합니다. 이 항목에서 단어는 주제의 확률 분포에 따라 가중치가 할당됩니다. LDA 사용자가 특정 단어가 주제에 얼마나 포함될 가능성을 말할 수 있는 것은 단어에 대한 확률적 할당입니다. 따라서 특정 주제에 할당된 단어 집합의 어휘적 관점에서 해당 주제가 실제로 무엇을 나타낼 수 있는지에 대한 통찰력을 얻을 수 있습니다.

표준 LDA 모델에는 모델을 실행하기 전에 염두에 두고 조정해야 하는 몇 가지 주요 파라미터가 있습니다.
1. n_components: 모형에 지정하는 주제의 수입니다.
2. α: 주제 우선순위에 연결할 수 있는 dirichlet 파라미터입니다.
3. β: 주제 우선순위에 연결된 dirichlet 파라미터입니다.

알고리즘을 호출하기 위해, 우리는 사이킷런의 LatentDirichletAllocation 함수를 통해 간단히 LDA 인스턴스를 만듭니다. 다양한 파라미터들의 값은 여러 종류의 검증을 통해 얻어진 것이 이상적입니다.

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
lda.fit(tf)
```
