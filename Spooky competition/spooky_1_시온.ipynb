먼저 다음과 같이 각각 Edgar Allen Poe, HP Lovecraft, Mary Shelley의 text를 저장하는 리스트를 생성한다.

```python
eap = train[train.author=='EAP']['text'].values
hpl = train[train.author=='HPL']['text'].values
mws = train[train.author=='MWS']['text'].values
```

하지만 일반적인 워드클라우드를 생성하는 것은 다소 지루하기 때문에, 사진을 가져오고 해당 그림의 윤곽을 워드클라우드의 마스크로 사용하는 방법을 소개한다. 따라서 선택한 사진은 작가의 가장 대표적이라고 생각하는 것들이다.

1. Edgar Allen Poe의 갈가마귀
2. HP Lovecraft의 Octopus Cthulu-thingy

먼저 사용할 이미지의 Base64 인코딩을 도출한 다음 특정 인코딩을 사용하고 이미지를 노트북에 다시 변환하다. 아래 셀은 사용할 세 이미지의 Base64 인코딩이 포함되어있다.

```python
eap_64 = b''
hpl_64 = b''
mws_64 = b''
```

이미지를 디코딩하기 위해 'codecs' 모듈을 임포트해 새 이미지로 저장한다. 이미지가 저장되면 다음과 같이 간단히 마스크로 로드할 수 있다.

```python
import codecs
# EAP의 마스크 생성
f1 = open('eap.png', 'wb')
f1.write(codecs.decode(eap_64,'base64'))
f1.close()
img1 = imread('eap.png')
hcmask = img1

f2 = open('mws.png', 'wb')
f2.write(codecs.decode(mws_64, 'base64'))
f2.close()
img2 = imread('mws.png')
hcmask2 = img2

f3 = open('hpl.png', 'wb')
f3.write(codecs.decode(hpl_64, 'base64'))
f3.close()
img3 = imread('hpl.png')
hcmask3 = img3
```

마지막으로 다음 몇 줄을 통해 워드클라우드를 시각화한다.

```python
# HP Lovecraft의 Cthulhu/squidy 의 워드클라우드
plt.figure(figsize=(16,13))
wc = WordCloud(background_color='black', max_words=10000,
              mask=hcmask3, stopwords=STOPWORDS, max_font_size=40)
wc.generate(' '.join(hpl))

plt.title('HP Lovecraft (Cthulhu-Squidy)', fontsize=20)
plt.imshow(wc.recolor(colormap='jet', random_state=17), alpha=0.98)
plt.axis('off')
```

```python
# Edgar Allen Poe의 the raven(갈가마귀) 워드클라우드
plt.figure(figsize=(20,18))
plt.subplot(211)
wc = WordCloud(background_color='black', max_words=10000, mask=hcmask,
              stopwords=STOPWORDS, max_font_size=40)
wc.generate(' '.join(eap))
plt.title('Edgar Allen Poe (The Raven)')
plt.imshow(wc.recolor(colormap='PuBu', random_state=17), alpha=0.9)
plt.axis('off')
```

```python
plt.figure(figsize=(20,18))
wc = WordCloud(background_color='black', max_words=10000,
              mask=hcmask2, stopwords=STOPWORDS, max_font_size=40)
wc.generate(' '.join(mws))
plt.title("Mary Shelley (Frankenstein's Monster)", fontsize=18)
plt.imshow(wc.recolor(colormap='viridis', random_state=17), alpha=0.9)
plt.axis('off')
```

# 2. Natural Language Processing

접하게 될 거의 모든 자연어처리(컴퓨터와 인간의 언어 간 상호작용을 탐구하는 분야) 작업 (토픽모델링, 워드 클러스터링, 문서 텍스트 분류 등)에서는 일반적으로 입력 원시 텍스트를 모델과 기계가 읽을 수 있는 형태로 변환하기 위해 이런 몇 가지 전처리 단계를 거쳐야한다. 랜덤 포레스트 모델에 단어 한 단락을 입력할 수 없으며 해당 단락이 어느 저자의 것인지 즉히 예측할 수 있다. 이면에서는 텍스트 전처리를 다음과 단계로 요약할 수 있다.
1. Tokenization : 텍스트를 개별 구성 단어로 구분
2. Stopwords : 발생 빈도가 너무 높으면 관련 텍스트를 탐지하는 데 도움이 되지 않으므로 너무 자주 발생하는 단어를 모두 삭제 (아주 드물게 발생하는 단어를 버리는 것도 고려)
3. Stemming : 동일한 의미를 전달하는 단일 상위 단어로 변형된 단어를 결합
4. Vectorization : 텍스트를 벡터 형식으로 변환. 가장 간단한 방법 중 하나는 유명한 bag-of-words 접근법. 여기서 행렬(말뭉치의 각 문서 또는 텍스트)를 만든다. 가장 간단한 형태에서, 이 행렬은 단어 빈도(단어 수)를 저장하며 종종 원시 텍스트의 벡터화라고 불린다.

Natural Language Toolkit(NLTK) : 자연어 처리를 보다 편리하게 하기 위해 NLP에서 가장 유용한 툴킷 중 하나인 NLTK 모듈 - 자연어처리 툴킷을 소개한다.
